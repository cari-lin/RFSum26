```python
import requests
import pandas as pd
```

```python
url = "https://bts-status.bicycletransit.workers.dev/phl"

data = requests.get(url).json()
```

```python
stations = pd.DataFrame([
    {
        "id": f["properties"]["id"],
        "name": f["properties"]["name"],
        "latitude": f["properties"]["latitude"],
        "longitude": f["properties"]["longitude"],
        "total_docks": f["properties"]["totalDocks"],
        "street_address": f["properties"]["addressStreet"]
    }
    for f in data["features"]
])
```

```python
stations.shape
```

```python
# add virtual station

stations.loc[len(stations)] = [
    3000,
    "Virtual Station",
    pd.NA,
    pd.NA,
    pd.NA,
    pd.NA
]
```

```python
stations.shape
```

```python
stations.dtypes
```

```python
stations["latitude"] = pd.to_numeric(
    stations["latitude"],
    errors="coerce"
)

stations["longitude"] = pd.to_numeric(
    stations["longitude"],
    errors="coerce"
)

stations["total_docks"] = pd.to_numeric(
    stations["total_docks"],
    errors="coerce"
).astype("Int64")
```

```python
stations.dtypes
```

```python
stations.to_csv("stations.csv", index=False)
```
