```python
import pandas as pd
```

```python
philly_weather = pd.read_csv('../../data/data/raw/philly_weather.csv')
```

```python
philly_weather.head()
```

```python
philly_weather = philly_weather.rename(columns={
    "STATION": "station",
    "NAME": "location",
    "LATITUDE": "latitude",
    "LONGITUDE": "longitude",
    "ELEVATION": "elevation",
    "DATE": "date",
    "AWND": "avg_wind",
    "PRCP": "precip",
    "SNOW": "snow",
    "SNWD": "snow_ground",
    "TMAX": "max_temp",
    "TMIN": "min_temp",
})
```

```python
philly_weather.head()
```

```python
philly_weather = philly_weather.drop(['station', 'location'], axis=1)
```

```python
philly_weather['date'] = pd.to_datetime(philly_weather['date'])
```

```python
philly_weather.head()
```

```python
weather = philly_weather.drop(['latitude', 'longitude', 'elevation'], axis=1)
```

```python
weather.head()
```

```python
weather.to_csv("weather.csv", index=False)
```
